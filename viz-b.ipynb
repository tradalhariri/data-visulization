```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
kc_houses = pd.read_csv('kc_house_data.csv')
```

```python
kc_houses.head()
```

```python
sns.set()
```

```python
sns.set_theme(style="dark")
houses_by_year = kc_houses.groupby('yr_built').sum()

sns.relplot(x="yr_built", y="price",hue='grade', data=houses_by_year)
```

### we conclude from the graph above that the prices of houses in USA inreases whenever the houses are newer and the house grade is high.

```python
sns.set_theme(style="dark")

sns.relplot(x="bathrooms", y="price",hue='bedrooms',palette='Set1', data=kc_houses)
```

### we conclude from the graph above that the prices of houses in USA inreases whenever the houses have more bathrooms and more bed rooms

```python
sns.scatterplot(x='bedrooms', y ='price' ,
data = kc_houses , hue = kc_houses['view'])
```

### we conclude from the graph above that the prices of houses in USA inreases whenever the houses have more bedrooms and the house has more views.

# Conclusion 

- The house prices in USA increases whenever the house was newer and the hous has high grade
- The house prices in USA increases whenever the house has more bathrooms and more bedrooms.
- The house prices in USA increases whenever the house has more bedrooms and more views.
